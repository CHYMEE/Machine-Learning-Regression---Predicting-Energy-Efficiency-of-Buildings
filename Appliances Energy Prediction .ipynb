```python
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import  r2_score,mean_absolute_error, mean_squared_error
%matplotlib inline
import warnings

warnings.filterwarnings("ignore")
```

```python
df_energy = pd.read_csv("energydata_complete.csv")
df_energy.head()
```

```python
#COLUMN RENAMING
column_rename = {'T1':'temp_kitchen', 'T2': 'temp_livingroom', 'T3':'temp_laundry', 'T4': 'temp_office', 
                 'T5':'temp_bathroom', 'T6': 'temp_outside_northside', 'T7':'temp_ironing', 'T8': 'temp_teenagerroom', 
                'RH_1':  'humidity_kitchen', 'RH_2': 'humidity_livingroom', 'RH_3': 'humidity_laundry',
                'RH_4': 'humidity_office', 'RH_5': 'humidity_bathroom', 
                'RH_6': 'humidity_outside', 'RH_7':  'humidity_ironing', 'RH_8': 'humidity_teenagerroom',
                'RH_9':  'humidity_parentroom', 'rv1': 'random_variable1','rv2': 'random_variable2',
                'T9':'temp_parentroom', 'T_out': 'temp_outside','RH_out':  'humidity_outside', }

df_energy = df_energy.rename(columns=column_rename)
df_energy
```

```python
#dropping the date and lights column
df_energy.drop(['lights','date'], axis=1,inplace=True)
df_energy.head(10)
```

```python
#normalizing
scalar = MinMaxScaler()

df_energy_normalize = pd.DataFrame(scalar.fit_transform(df_energy), columns=df_energy.columns)
df_energy_normalize
```

```python
#splitting the data into independent(X) and target(y) variables
X = df_energy_normalize.drop(['Appliances'], axis=1)
y = df_energy_normalize['Appliances']
```

```python
X
```

```python
#train-test splitting
x_train, x_test, y_train, y_test= train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
x_train.shape,  x_test.shape, y_train.shape, y_test.shape
```

```python
#fit the model to the training dataset
linear_model = LinearRegression()
lin = linear_model.fit(x_train, y_train)
lin
```

```python
#evaluating model on test set
predicted_values = linear_model.predict(x_test)
predicted_values
```

# 1. MAE

```python
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 3)
mae
```

```python
#Getting the features with the lowest and highest weights

def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df
model_weight = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
model_weight
```

```python
ridge_reg = Ridge(alpha = 0.4)
ridge_reg.fit(x_train, y_train)
prediction = ridge_reg.predict(x_test)
ridge_rmse = np.sqrt(mean_squared_error(y_test, prediction))
round(ridge_rmse,3)
```
